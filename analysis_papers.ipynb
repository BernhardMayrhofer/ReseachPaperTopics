```r
library("ggplot2")
library("reshape2")
```

```r
gamma_ = read.table("gamma.dat")[1:50,]
papers_ = data.frame(read.csv("papers.csv", header = FALSE))[1:50,]
research_ = cbind(papers_[1],gamma_)
colnames(research_) <- c('paper','Topic 1','Topic 2','Topic 3','Topic 4','Topic 5','Topic 6','Topic 7','Topic 8','Topic 9','Topic 10')
```

```r
research_temp_ <- melt(research_, id = c("paper"))

heatmap.plot <- ggplot(data = research_temp_, aes(x = variable, y = paper)) +
  geom_tile(aes(fill = value)) +
  scale_fill_gradient2() +
  theme(axis.text.y = element_text(size = 6))

# Preview the heatmap
print(heatmap.plot)
```
